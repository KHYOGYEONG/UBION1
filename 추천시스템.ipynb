### Content Based Filtering

### 목적 : 영화마다 코사인 유사도를 구하기 위한 작업임

    - 숫자로 바꾸는 작업

```python
import pandas as pd

movies = pd.read_csv('./datasets/tmdb_5000_movies.csv')
movies.head()
```

```python
import ast
ast.literal_eval(movies['genres'][0]) # name인 장르만 가져오기 위해 확인하고 {하나씩} 아래 함수에 넣어줌
```

```python
def value_import(x):
    return [i['name'] for i in ast.literal_eval(x)]
movies['genres'] = movies['genres'].apply(value_import)
movies['keywords'] = movies['keywords'].apply(value_import)
```

```python
movies['genres'][0]
```

```python
movies['keywords'][0]
```

    - 장르와 키워드 기반으로 유사한 걸 뽑기 위해 가져온거임

```python
# 장르와 키워드를 합치기
# soup이라는 컬럼에 넣어주기
movies['soup'] = movies['genres'] + movies['keywords']
```

```python
movies.head()
```

```python
movies['soup']
```

    - 콤마(,), 리스트를 없애야함

```python
def join_words(x):
    return (' ').join(x)

movies['soup_literal'] = movies['soup'].apply(join_words)
```

```python
movies['soup_literal'][0]
```

    - 이제 숫자로 바꿀거임
    - 방법은 Count, TF-IDF가 있음
    - count : 카운트 기반 벡터화, 중요한 의미가 없는 것도 빈도수가 많으면 중요하게 본다는 단점이 있음
    - TF-IDF 벡터화 : 위 단점을 극복한것임. ex_ 신문기사, 일기 등

## sklearn.feature_extraction.text.CountVectorize
- class sklearn.feature_extraction.text.CountVectorizer(*, input='content', encoding='utf-8', decode_error='strict', strip_accents=None, lowercase=True, preprocessor=None, tokenizer=None, stop_words=None, token_pattern='(?u)\b\w\w+\b', ngram_range=(1, 1), analyzer='word', max_df=1.0, min_df=1, max_features=None, vocabulary=None, binary=False, dtype=<class 'numpy.int64'>)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

count_vector = CountVectorizer()
soup_mat = count_vector.fit_transform(movies['soup_literal'])
```

```python
soup_mat.shape # 7164개인 벡터로 만들어진거임
```

    - 각 영화마다 코사인 유사도(cosine_similarity)를 구할거임
    - 라이브러리 안에 있음

### sklearn.metrics.pairwise.cosine_similarity
- sklearn.metrics.pairwise.cosine_similarity(X, Y=None, dense_output=True)

```python
from sklearn.metrics.pairwise import cosine_similarity

soup_similarity = cosine_similarity(soup_mat, soup_mat)
```

```python
soup_similarity[0]
```

    - 유사도를 구했으니
    - soup_similarity(유사도)를 영화ID를 기준으로 내림차순 하기

```python
soup_sim_sorted_ind = soup_similarity.argsort()[:, ::-1] # 내림차순 정렬하는 방법임
# argsort()는 큰 것부터 인덱스 번호를 찍어줌
soup_sim_sorted_ind # 유사도가 큰 거의 자리를 알려주는 거임. 몇번쨰 있는지지
```

    - 다른 방법 NearestNeighbors를 사용해서 유사도를 구할거임

### sklearn.neighbors.NearestNeighbors
- class sklearn.neighbors.NearestNeighbors(*, n_neighbors=5, radius=1.0, algorithm='auto', leaf_size=30, metric='minkowski', p=2, metric_params=None, n_jobs=None)
    - n_neighbors=5 가장 가까운 5개를 찾아줌
    - metric='minkowski' 어떤 거리? 우리는 코사인!

```python
# NearestNeighbors 를 활용한 유사도
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(n_neighbors=20, metric='cosine')
model_knn.fit(soup_mat)
```

    - NearestNeighbors안에 있는 메소드(kneighbors) 사용
    - kneighbors([X, n_neighbors]) : 지정한 것(X)에서 몇개(n_neighbors)찾아줌

```python
similariity, indices_result = model_knn.kneighbors(soup_mat[1], n_neighbors=11)
```

```python
indices_result[0][1: ] # 자기자신을 제외한 것만 보기
```

```python
movies['original_title'][indices_result[0][1:]] # 영화 이름 확인
```
